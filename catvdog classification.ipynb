```python
X_train=np.loadtxt('D:/kaggle/Image Classification CNN Keras Dataset/input.csv',delimiter=',')
```

```python
y_train=np.loadtxt('D:/kaggle/Image Classification CNN Keras Dataset/labels.csv',delimiter=',')
X_test=np.loadtxt('D:/kaggle/Image Classification CNN Keras Dataset/input_test.csv',delimiter=',')
y_test=np.loadtxt('D:/kaggle/Image Classification CNN Keras Dataset/labels_test.csv',delimiter=',')
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_test.shape)
```

```python
print(y_train.shape)
```

```python
X_train=X_train.reshape(len(X_train),100,100,3)
X_test=X_test.reshape(len(X_test),100,100,3)
y_test=y_test.reshape(len(y_test),1)
y_train=y_train.reshape(len(y_train),1)
```

```python
X_train[1,:]
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
import matplotlib.pyplot as plt
```

```python
idx = np.random.randint(0, len(X_train))
plt.imshow(X_train[idx])


```

```python
model = Sequential()
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(100,100,3)))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(32, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=64)
```

```python
model.evaluate(X_test,y_test)
```
